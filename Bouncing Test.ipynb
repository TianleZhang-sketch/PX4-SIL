```python
import scipy.integrate
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
g = 9.8
k = 1000
def y(t, x):
    if x[0] <0:
        a = -g -k*x[0]
    else:
        a = -g
    xdot = [x[1], a]
    return xdot
ode = scipy.integrate.ode(y)
ode.set_initial_value([1,0], 0)
y = []
t = []
dt = 0.001
tf = 10
n_t = int(tf/dt)
y = np.zeros((n_t,2))
t = np.zeros(n_t)
count = 0
while ode.t + dt< 10:
    ode.integrate(ode.t + dt)
    y[count] = ode.y
    t[count] = ode.t
    count = count + 1
plt.plot(t,y)
```

Calculate the energy to show that the numerical scheme isn't violating the conversation of energy.

```python
x = y[:,0]
x_dot = y[:,1]
e = x_dot**2/2 + 1*9.8*x + np.where(x < 0, 1, 0)*k*x**2/2
```
